```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def negative(to_convert_image):
    img = 255 - to_convert_image

    return img
```

```python
image = plt.imread('real-hippo.jpeg')
def cartoonify(image):    
    img_blur = cv2.blur(image, (5, 5)) # blur image

    edges = cv2.Canny(img_blur, 100, 100) # find edges of image through thresholding, smoothing, etc

    invert = negative(edges) # invert image

    plt.title('Cartoonified image')
    plt.imshow(invert, cmap='gray')
    plt.show()

    cv2.imwrite('cartoonified_bozu.png', invert)

cartoonify(image)
```
